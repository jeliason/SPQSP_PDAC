```python
import os
import pandas as pd
import pyvista as pv
import matplotlib.pyplot as plt
import numpy as np
```

```python
abm_path = os.path.dirname(os.getcwd()) + "/PDAC/sim/HCC_outputs/snapShots/"

abm_files = sorted(os.listdir(abm_path))
```

```python
df = pd.read_csv(abm_path + abm_files[0], header = 0)
```

```python
points = df[['x','y','z']].to_numpy().astype(np.float32)
clouds = pv.PolyData(points)
clouds['state'] = df['Type']
```

```python
p = pv.Plotter()
p.add_mesh(clouds,scalars = 'state', point_size=10,render_points_as_spheres=True, label = 'state')
p.add_axes()
p.view_xy()
p.add_legend()
p.show()
```
